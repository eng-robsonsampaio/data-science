```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#How to work with seaborn sample data frame
dataframe = sns.load_dataset('tips')
```

```python
dataframe.head(n=5)
```

```python
dataframe.shape
```

```python
dataframe.corr(numeric_only=True)
```

```python
dataframe.describe()
```

```python
sns.lmplot(data=dataframe, 
           x="total_bill", 
           y="tip",
           hue='sex')
```

```python
sns.jointplot(data=dataframe, x='total_bill', y='tip', kind='reg')
```

```python
sns.lmplot(data=dataframe, x='total_bill', y='tip', hue='smoker')
```

```python
# color.adobe.com
sns.relplot(x='total_bill', 
            y='tip', 
            col= 'time', 
            hue='smoker', 
            style='smoker', 
            size='size', 
            data=dataframe,
            palette=['#EB6946', '#CE2FEB'])
```

```python
fig, axes = plt.subplots(1,2, sharey=True)
sns.boxplot(data=dataframe, x='day', y='tip', ax=axes[0])
sns.scatterplot(data=dataframe, x='total_bill', y='tip', ax=axes[1])
```
