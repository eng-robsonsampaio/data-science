```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
vetor_bidimensional = [[-2,1], [0,-3], [1,4], [1,1], [4,4], [-2,-1]]
vetor_bidimensional
```

```python
X = np.array(vetor_bidimensional)

# Escolher o número de clusters (2 clusters)
num_clusters = 2

# Definir manualmente os dois primeiros centros
centros_iniciais = np.array([[1, 4], [4, 4]])

# Criar o modelo K-Means com os centros iniciais definidos
kmeans = KMeans(n_clusters=num_clusters, init=centros_iniciais)

# Treinar o modelo com o conjunto de dados
kmeans.fit(X)

# Obter as coordenadas dos centroides encontrados pelo K-Means
centroids = kmeans.cluster_centers_

# Obter as etiquetas dos clusters para cada ponto no conjunto de dados
labels = kmeans.labels_

# Plotar os pontos no gráfico de dispersão, com cores diferentes para cada cluster
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')

# Plotar os centroides em vermelho
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)
plt.title('K-Means Clustering com 2 clusters e Centros Iniciais Definidos Manualmente')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid()
```
