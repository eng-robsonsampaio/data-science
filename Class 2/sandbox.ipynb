```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o conjunto de dados "tips" do Seaborn
tips = sns.load_dataset("tips")

# Plotando um histograma usando histplot() e alterando o nome na legenda
ax = sns.histplot(data=tips, x="total_bill", hue="sex")

# Definindo o rótulo da legenda
ax.legend(labels=["Feminino", "Masculino"])

# Exibindo o gráfico
plt.show()
```
