```python
def gera_lista(valor_min: float , valor_max: float, 
               quant_total: int, casa_dec:int) -> list:
  """
    Cria uma lista aleátoria de N números com base em valores
    minimo e máximo.
    Arguments:
        valor_min: an float
        valor_max: an float
        quant_total: an integer
        casa_dec: an integer
    Returns:
        Uma lista com quant_total de números aleátorios.
    """

  import random
  
  random_floats = [round(random.uniform(valor_min, valor_max), 
                         casa_dec) for _ in range(quant_total)]

  return random_floats
```

```python
valor_min_bill_lenght = df_pinguin['bill_length_mm'].min()
valor_min_bill_depth = df_pinguin['bill_depth_mm'].min()
valor_min_flipper = df_pinguin['flipper_length_mm'].min()
valor_min_mass = df_pinguin['body_mass_g'].min()
```

```python
valor_max_bill_lenght = df_pinguin['bill_length_mm'].max()
valor_max_bill_depth = df_pinguin['bill_depth_mm'].max()
valor_max_flipper = df_pinguin['flipper_length_mm'].max()
valor_max_mass = df_pinguin['body_mass_g'].max()
```

```python
lis_bill_lenght = gera_lista(valor_min_bill_lenght, valor_max_bill_lenght, 4500, 2)
lis_bill_depth = gera_lista(valor_min_bill_depth, valor_max_bill_depth, 4500, 2)
lis_flipper = gera_lista(valor_min_flipper, valor_max_flipper, 4500, 2)
lis_mass = gera_lista(valor_min_mass, valor_max_mass, 4500, 2)
```

```python
df_novo = pd.DataFrame(columns=nome_das_colunas)
```
